<p>YouTube is a website designed for users to share videos with everyone. In addition, users can watch the video as entertainment. Statistics show that every 35 hours of video is uploaded to YouTube. YouTube was created in 2005 by Chad Hurley, Jawed Karim, and Steve Chen. The main purpose is to broadcast and share video content. But since then it is also become both an archive for storing favourite clips, songs, and jokes, as well as a marketing site for companies to promote their products. Film and TV organizations keep up a tight control over their own content and block illegal sharing of their product. However, several now use YouTube to permit fans to see high-quality trailers or repeats of recently aired shows.
Nowadays the term 'viral video' is normal. This refers to a video clip that individuals have enjoyed and like so much that they have shared its link by email to a huge number of other people around the world. basically, it is spread like an infection. Organizations have understood that they can harness this ability to reach potential clients and have made their own YouTube account posting ads and other advertising recordings.
